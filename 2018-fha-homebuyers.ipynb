```python
import pandas as pd
```

```python
# From the data dictionary (2018_Single_Family_National_File_A.pdf)
col_names = [
         'enterprise',
         'record',
         'metro',
         'tract_pct_minority',
         'tract_income_ratio',
         'borrower_income_ratio',
         'ltv',
         'loan_purpose',
         'fed_guarantee',
         'borrower_race',
         'borrower_gender',
         'coborrower_race',
         'coborrower gender',
         'units',
         'affordability'
        ]
```

### Freddie Mac

```python
temp1 = 'data/fhlmc_sf2018a_loans.txt'
```

```python
df1 = pd.read_csv(temp1,
                 sep="\s+", #or delim_whitespace=True, #separator is whitespace
                 header=None, #no header
                 names=col_names,
                 index_col=1
                )
df1.head()
```

## Fannie Mae

```python
temp2 = 'data/fnma_sf2018a_loans.txt'
```

```python
df2 = pd.read_csv(temp2,
                 sep="\s+", #or delim_whitespace=True, #separator is whitespace
                 header=None, #no header
                 names=col_names,
                 index_col=1
                )
df2.head()
```

## Combine files
Single dataset with manageable sample size

```python
df1=df1.sample(5000)
```

```python
df2=df2.sample(5000)
```

```python
df3=df1.append(df2)
df3.shape
```

```python
df3.head()
```

```python
df4=df3.reset_index()
df4.head()
```

```python
df4.to_csv('data/mortgage_sample.csv', index=False)
```
